```python
import numpy as np
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('default')
from scipy.integrate import simps


import numpy as np
from scipy import stats
import pandas as pd
from tqdm.notebook import tqdm
from twobitreader import TwoBitFile
import math
from scipy.signal import savgol_filter

def sequence(reads):
    sequences = []
    genome = TwoBitFile("hg38.2bit")
    chroms = reads.column("chrom-RBP")
    starts = reads.column("start-RBP")
    stops = reads.column("stop-RBP")
    strands = reads.column("strand")
    for i in (np.arange(reads.num_rows)):
        chrom = chroms[i]
        start = int(starts[i])
        stop = int(stops[i])
        strand = strands[i]
        chromid = genome[chrom]
        sequence = chromid[start-1:stop].lower().replace("t", "u")
        if strand == "-":
            sequence = complement(sequence)
        sequences.append(sequence)
    reads = reads.with_column("RNA-Seq-Motif", sequences)
    return reads
def complement(dna):
    complement = {'a': 'u', 'c': 'g', 'g': 'c', 'u': 'a' ,'n':'n'}
    return ''.join([complement[base] for base in dna[::-1]])

```

```python
for dir_n in ['E_STIM_vs_UN_3pUTRs', 'E_UN_vs_STIM_3pUTRs']:
    for RBP in ['PABPC1.bed.txt','PABPC4.bed.txt']:
        
        data = pd.read_csv(dir_n + '/'+RBP, sep="\t", header=None)
        data.columns = ["chrom", "start", "stop", "geneid","DBE",'strand','RNA-Seq','chrom-RBP','start-RBP','stop-RBP','RBP']
        data =Table.from_df(data)
        data = data.with_column('motif_relative_pos', [round(((a+b)/2) - ((c+d)/2)) for a,b,c,d in zip(data['start'],data['stop'],data['start-RBP'],data['stop-RBP'])])
        data = data.with_column('geneid', [i.split('-')[0] for i in data.column('geneid')])
        data = sequence(data)
        
        if dir_n == 'E_STIM_vs_UN_3pUTRs':
            tbl_to_use = stim_tbl
        if dir_n == 'E_UN_vs_STIM_3pUTRs':
            tbl_to_use = unstim_tbl
        tbl_to_use = tbl_to_use.select('Differential Binding Enrichment(DBE)','percentile' ).relabeled('Differential Binding Enrichment(DBE)','DBE')
        tbl_to_use = tbl_to_use.with_column('DBE', np.round(tbl_to_use['DBE'], 2)).with_column('percentile', np.round(tbl_to_use['percentile'], 2))
        tbl_to_use = tbl_to_use.group(0, np.mean).relabeled('percentile mean', 'percentile')
        
        data = data.with_column('DBE', np.round(data['DBE'], 2))
        data = data.join('DBE',tbl_to_use).sort('percentile', descending=True).select(1,2,3,4,0,13,5,6,7,8,9,10,11,12)
        data.to_csv((dir_n + '-'+RBP).replace('.bed.txt','.csv'))
```

```python
import os
from scipy.stats import percentileofscore
from tqdm.notebook import tqdm




def q_C_list(cond_2,DBE_col):
    arr = cond_2.column(DBE_col)

    # pre-sort array
    arr_sorted =  sorted(arr)

    # calculate percentiles using scipy func percentileofscore on each array element
    s = pd.Series(arr)
    percentiles = s.apply(lambda x: percentileofscore(arr_sorted, x))
    def quality_score(pct):
        if pct <= 25:
            return 1
        if pct <= 50:
            return 2
        if pct <=70:
            return 3
        if pct <= 80:
            return 4
        if pct <= 85:
            return 5 
        if pct <= 90:
            return 6 
        if pct <= 95:
            return 7
        if pct <= 97.5:
            return 8 
        if pct <= 99:
            return 9
        if pct <= 100:
            return 10
    cond_2 = cond_2.with_columns("percentile", percentiles)
    return cond_2



def concat_files(files):
    starts, stops, strands, chroms, geneIDs, depths = [], [], [], [], [], []
    for file in files:
        tbl = Table().read_table("Output/" + file)
        [starts.append(i) for i in tbl.column("start")]
        [stops.append(i) for i in tbl.column("stop")]
        [strands.append(i) for i in tbl.column("strands")]
        [chroms.append(i) for i in tbl.column("chrom")]
        [geneIDs.append(i) for i in tbl.column("geneid")]
        [depths.append(i) for i in tbl.column("AUC Differential Binding")]
    
        
    tbl = Table().with_columns("chrom", chroms,"start", starts, "stop", stops,"geneid",geneIDs,"Differential Binding Enrichment(DBE)",depths, "strand", strands )
    tbl = sequence(tbl)
    tbl = q_C_list(tbl, 'Differential Binding Enrichment(DBE)')
    return tbl

```

```python
stim_tbl = q_C_list(Table().read_table('E_STIM_vs_UN_3pUTRs.csv'),'Differential Binding Enrichment(DBE)')
```

```python
unstim_tbl = q_C_list(Table().read_table('E_UN_vs_STIM_3pUTRs.csv'),'Differential Binding Enrichment(DBE)')
```
