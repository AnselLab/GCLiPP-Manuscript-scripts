```python
import numpy as np
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('default')
import warnings
warnings.simplefilter('ignore', FutureWarning)

import numpy as np
from scipy import stats
import subprocess
import pandas as pd
import time
from tqdm import tqdm_notebook as tqdm


import requests
from lxml import etree
from lxml.etree import fromstring
from twobitreader import TwoBitFile


def get_sequence(chrom,start, stop):

    req = requests.get(("http://genome.ucsc.edu/cgi-bin/das/mm10/dna?segment=" + chrom + ":" + str(int(start)) + "," + str(int(stop))), stream=True)
    req.raw.decode_content = True  # ensure transfer encoding is honoured
    b = etree.parse(req.raw)
    return "".join([i[1:] for i in str(etree.tostring(b)).split("length")[1].split("\\n</DNA>\\")[0].split(">")[1].split("\\")[1:]])
def sequence(reads):
    sequences = []
    genome = TwoBitFile("hg38.2bit")
    chroms = reads.column("chrom")
    starts = reads.column("start")
    stops = reads.column("stop")
    strands = reads.column("strand")
    for i in (np.arange(reads.num_rows)):
        chrom = chroms[i]
        start = int(starts[i])
        stop = int(stops[i])
        strand = strands[i]
        chromid = genome[chrom]
        sequence = chromid[start-1:stop].lower().replace("t", "u")
        if strand == "-":
            sequence = complement(sequence)
        sequences.append(sequence)
    reads = reads.with_column("RNA sequences", sequences)
    return reads
def complement(dna):
    complement = {'a': 'u', 'c': 'g', 'g': 'c', 'u': 'a' ,'n':'n'}
    return ''.join([complement[base] for base in dna[::-1]])




```

```python
from scipy.stats import ks_2samp

def cdf(x, xmin,xmax,*args, **kwargs):
    proportion = []
    for x_i in np.linspace(xmin,xmax,10000):
        proportion.append((sum(x<=x_i)/len(x)))
    plt.plot(np.linspace(xmin,xmax,10000), proportion, *args, **kwargs)
    return np.asarray(proportion)
```
